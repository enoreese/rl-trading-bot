```python
from tensortrade.feed.core import Stream, DataFeed, NameSpace
from tensortrade.oms.instruments import USD, BTC, ETH, LTC
from tensortrade.oms.wallets import Wallet, Portfolio
from tensortrade.oms.exchanges import Exchange
from tensortrade.oms.services.execution.simulated import execute_order

import ta
import tensortrade.env.default as default
```

```python
def rsi(price: Stream[float], period: float) -> Stream[float]:
    r = price.diff()
    upside = r.clamp_min(0).abs()
    downside = r.clamp_max(0).abs()
    rs = upside.ewm(alpha=1 / period).mean() / downside.ewm(alpha=1 / period).mean()
    return 100*(1 - (1 + rs) ** -1)


def macd(price: Stream[float], fast: float, slow: float, signal: float) -> Stream[float]:
    fm = price.ewm(span=fast, adjust=False).mean()
    sm = price.ewm(span=slow, adjust=False).mean()
    md = fm - sm
    signal = md - md.ewm(span=signal, adjust=False).mean()
    return signal
```

```python
allowed_cols = ["open","high","low","close","volume"]
def create_env(window_size):
    bitfinex_btc = cdd.fetch("Bitfinex", "USD", "BTC", "1h")
    bitfinex_eth = cdd.fetch("Bitfinex", "USD", "ETH", "1h")
    bitfinex_ltc = cdd.fetch("Bitfinex", "USD", "LTC", "1h")
    
    bitfinex_btc.to_csv("data/bitfinex_btc_1h.csv")
    bitfinex_eth.to_csv("data/bitfinex_eth_1h.csv")
    bitfinex_ltc.to_csv("data/bitfinex_ltc_1h.csv")

    btc_close = Stream.source(list(bitfinex_btc["close"]), dtype="float").rename("BTC_close")
    btc_volume = Stream.source(list(bitfinex_btc["volume"]), dtype="float").rename("BTC_volume")
    btc_open = Stream.source(list(bitfinex_btc["open"]), dtype="float").rename("BTC_open")

    eth_close = Stream.source(list(bitfinex_eth["close"]), dtype="float").rename("ETH_close")
    eth_volume = Stream.source(list(bitfinex_eth["volume"]), dtype="float").rename("ETH_volume")
    eth_open = Stream.source(list(bitfinex_eth["open"]), dtype="float").rename("ETH_open")

    ltc_close = Stream.source(list(bitfinex_ltc["close"]), dtype="float").rename("LTC_close")
    ltc_volume = Stream.source(list(bitfinex_ltc["volume"]), dtype="float").rename("LTC_volume")
    ltc_open = Stream.source(list(bitfinex_ltc["open"]), dtype="float").rename("LTC_open")

    features = []
    for df, label in zip([bitfinex_btc, bitfinex_eth, bitfinex_ltc], ["btc","eth","ltc"]):
        for c in df.columns[1:]:
            s = Stream.source(list(df[c]), dtype="float").rename(f"{label}_{df[c].name}")
            features += [s]

#     cp = Stream.select(features, lambda s: s.name == "btc_close")

    features = [
        btc_close,
        btc_volume,
        btc_close.rolling(window=10).mean().rename("btc_fast"),
        btc_close.rolling(window=50).mean().rename("btc_medium"),
        btc_close.rolling(window=100).mean().rename("btc_slow"),
        btc_close.log().diff().fillna(0).rename("btc_lr"),
        rsi(btc_close, period=20).rename("btc_rsi"),
        macd(btc_close, fast=10, slow=50, signal=5).rename("btc_macd"),

        ltc_close,
        ltc_volume,
        ltc_close.rolling(window=10).mean().rename("ltc_fast"),
        ltc_close.rolling(window=50).mean().rename("ltc_medium"),
        ltc_close.rolling(window=100).mean().rename("ltc_slow"),
        ltc_close.log().diff().fillna(0).rename("ltc_lr"),
        rsi(ltc_close, period=20).rename("ltd_rsi"),
        macd(ltc_close, fast=10, slow=50, signal=5).rename("ltd_macd"),

        eth_close,
        eth_volume,
        eth_close.rolling(window=10).mean().rename("eth_fast"),
        eth_close.rolling(window=50).mean().rename("eth_medium"),
        eth_close.rolling(window=100).mean().rename("eth_slow"),
        eth_close.log().diff().fillna(0).rename("eth_lr"),
        rsi(eth_close, period=20).rename("eth_rsi"),
        macd(eth_close, fast=10, slow=50, signal=5).rename("eth_macd")
    ]

    feed = DataFeed(features)
    feed.compile()
    
    bitfinex = Exchange("bitfinex", service=execute_order)(
        Stream.source(list(bitfinex_btc["close"]), dtype="float").rename("USD-BTC"),
        Stream.source(list(bitfinex_eth["close"]), dtype="float").rename("USD-ETH"),
        Stream.source(list(bitfinex_ltc["close"]), dtype="float").rename("USD-LTC"),
    )

    portfolio = Portfolio(USD, [
        Wallet(bitfinex, 100 * USD),
        Wallet(bitfinex, 0 * BTC),
        Wallet(bitfinex, 0 * LTC),
        Wallet(bitfinex, 0 * ETH)
    ])


    renderer_feed = DataFeed([
        Stream.source(list(bitfinex_btc["date"])).rename("date"),
        Stream.source(list(bitfinex_btc["close"]), dtype="float").rename("btc_close"),
        Stream.source(list(bitfinex_eth["close"]), dtype="float").rename("eth_close"),
        Stream.source(list(bitfinex_ltc["close"]), dtype="float").rename("ltc_close"),
#         Stream.source(list(data["close"]), dtype="float").rename("close"),
#         Stream.source(list(data["volume"]), dtype="float").rename("volume")
    ])

    reward_scheme = default.rewards.RiskAdjustedReturns()
    action_scheme = default.actions.ManagedRiskOrders(
        stop=[0.02, 0.04, 0.06],
        take=[0.02, 0.03],
        trade_sizes=4,
        trade_type='market',
        durations=[30, 60],
    )


    env = default.create(
        portfolio=portfolio,
        action_scheme=action_scheme,
        reward_scheme=reward_scheme,
        feed=feed,
    #     renderer_feed=renderer_feed,
        renderer=default.renderers.ScreenLogger(),
        window_size=window_size
    )
    return env, portfolio
```

```python
env, portfolio = create_env(20)
```
